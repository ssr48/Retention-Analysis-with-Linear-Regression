# Decision Trees

```python
# create a regressor object 
regressor = DecisionTreeRegressor(random_state = 0)  
  
# fit the regressor with X and Y data 
regressor.fit(rf_feat, rf_label) 

# test the output 
predictions = regressor.predict(test_feat_rf) 

# Calculate the absolute errors
errors = abs(predictions - test_label_rf)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
print('Difference in Mean: ', np.mean(test_label_rf) - np.mean(predictions))
print('RMSE: ', sqrt(mean_squared_error(test_label_rf, predictions)))
```

# KNN Means

```python
#Standardize
scaler = StandardScaler()
train_scaled = scaler.fit_transform(rf_feat)
test_scaled = scaler.transform(test_feat_rf)

#Fit Model
model = KNeighborsRegressor()
model.fit(train_scaled, rf_label)

#Check Metrics on actual data
mse = mean_squared_error(rf_label, model.predict(train_scaled))
mae = mean_absolute_error(rf_label, model.predict(train_scaled))
print("Train Data: mse = ",mse," & mae = ",mae," & rmse = ", sqrt(mse))

#Check metrics on test data
test_mse = mean_squared_error(test_label_rf, model.predict(test_scaled))
test_mae = mean_absolute_error(test_label_rf, model.predict(test_scaled))
print("Test Data: mse = ",test_mse," & mae = ",test_mae," & rmse = ", sqrt(test_mse))
```
